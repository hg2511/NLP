```python
import gensim
```

```python
from nltk import word_tokenize,pos_tag
from nltk.corpus import wordnet as wn
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
stop_words = set(stopwords.words('english'))
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
sen = []
for i in samp_list:
    token = word_tokenize(i)
    token = [w.lower() for w in token]
    words = [word for word in token if word.isalpha()]
    words = [w for w in words if not w in stop_words]
    words = [lemmatizer.lemmatize(w) for w in words]
    sen.append(words)
```

```python
sen
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(sen)]
```

```python
model = Doc2Vec(documents, vector_size=100)
```

```python
model.n_similarity(sen[0],sen[1])
```

```python
score = 1-model.n_similarity(sen[0],sen[1])
```
